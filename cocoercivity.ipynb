```python
import torch

array = torch.load('cocoercivity_100.pt')
print(array.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np

array_for_plot = array[:,:80].flatten()
min_value = torch.min(array_for_plot)
logbins = np.logspace(torch.log10(min_value),0,10)
# 새로운 Figure 객체 생성하여 크기 지정
plt.figure(figsize=(4, 3))
plt.hist(array_for_plot, bins=logbins, color='blue', edgecolor='black')
#plt.xlabel(r'$\langle \mathcal{E} \mathcal{D} z - \mathcal{E} \mathcal{D} z, z - z \rangle$')
#plt.ylabel('Frequency')
#plt.title('Histogram')
plt.xscale('log')

plt.axvline(x=min_value, color='red', linestyle='-')
plt.text(0.2, 0.7, r'$\beta=\mathrm{{{:.2e}}}$'.format(min_value), fontsize=12, transform=plt.gcf().transFigure)


plt.show()
```
