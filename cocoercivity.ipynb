```python
import torch

array = torch.load('vanila_process_20240503-195636.pt')
print(array.shape)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FuncFormatter

array_for_plot = array[:,:80].flatten()
min_value = torch.min(array_for_plot)
percentile = np.percentile(array_for_plot, 0.1)
logbins = np.logspace(torch.log10(min_value),0,10)
# 새로운 Figure 객체 생성하여 크기 지정
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(4, 3))
plt.hist(array_for_plot, bins=logbins, color='gray', edgecolor='black')
#plt.xlabel(r'$\langle \mathcal{E} \mathcal{D} z - \mathcal{E} \mathcal{D} z, z - z \rangle$')
#plt.ylabel('Frequency')
#plt.title('Histogram')
plt.xscale('log')

# yticks의 레이블을 지정하는 함수 정의
def format_yticks(value, pos):
    return '{:.0f}k'.format(value / 1000)

# yticks에 함수 적용
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_yticks))

plt.axvline(x=min_value, color='red', linestyle='-')
plt.axvline(x=percentile, color='blue', linestyle='--')
plt.text(0.18, 0.80, r'$\min=$', fontsize=14, transform=plt.gcf().transFigure)
plt.text(0.18, 0.73, r'${:.1f}\times10^{{-5}}$'.format(min_value* 1e5), fontsize=14, transform=plt.gcf().transFigure)
plt.text(0.5, 0.80, 'percentile(0.1)=', fontsize=11, transform=plt.gcf().transFigure)
plt.text(0.5, 0.73, r'${:.1f}\times10^{{-3}}$'.format(percentile*1e3), fontsize=13, transform=plt.gcf().transFigure)

plt.tight_layout()
plt.savefig('/home/icl2/khlee/InstaFlow/vanila_inf.pdf')
plt.show()
```
